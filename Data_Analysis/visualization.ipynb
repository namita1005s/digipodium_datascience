# Matplotlib

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in 2D and 3D.
 It provides a wide range of customizable plotting options and is widely used in data analysis, scientific research, and various other fields.

```python
import numpy as np
import matplotlib.pyplot as plt # this is used for plot the graph
import matplotlib as mpl # used for plot styling
```

```python
#Here's a simple example of creating a basic line plot:
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.figure(figsize=(4, 2))
# Create a line plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')

# Display the plot
plt.show()
```

In Matplotlib, a "figure" is like that piece of paper, and an "axis" represents each small plot you draw on the paper.

```python
# Create a figure and a single set of axes
# Create a figure (fig) with one subplot (ax) and set the figure size to (5, 2)
fig, ax = plt.subplots(figsize=(5,2)) 
#data
x = [1,2,3,4,5]
y = [3,4,4,5,3]
# Plot the data on the subplot (ax) using the 'o-' format string
ax.plot(x,y, 'o-')
```

When you use plt.plot(x, y) without explicitly specifying any axes, Matplotlib implicitly creates a subplot for you to hold the plot. In this case, the entire figure (fig) serves as the default subplot to contain the plot.
This behavior is convenient for simple plots when you don't require complex layout arrangements with multiple subplots. When you only need a single plot, you can directly use plt.plot() after defining the figure size, and Matplotlib will handle the creation of a single subplot for you.

```python
fig = plt.figure(figsize=(3, 3))
plt.plot(x,y)
'''By "without explicitly specifying any axes," I mean that you did not use plt.subplots() or plt.add_subplot() to create a specific set of axes to draw the plot.
When you use plt.plot(x, y) directly after creating a figure with plt.figure(figsize=(3, 3)), Matplotlib implicitly creates a set of axes for you.
In this case, the figure (fig) itself becomes the default axes for the plot.
So, when you call plt.plot(x, y), Matplotlib automatically uses the figure (fig) as the axes to display the line plot.'''
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
#the creation of the figure (fig) and axes (subplots) array (axs) is implicit.
```

implicit vs explicit graphs

```python


# Define the x values from -5 to 5 with a step of 0.2
x = np.arange(-5, 5, 0.2)

# Calculate the y values for sin(x), cos(x), and tan(x)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Create a figure with a 1x3 grid of subplots and set the figure size to (15, 3)
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15, 3))

# Plotting sin(x) on the first subplot (the leftmost subplot)
axs[0].plot(x, y1, 'r--')
axs[0].set_title('sin(x)')

# Plotting cos(x) on the second subplot (the middle subplot)
axs[1].plot(x, y2, 'b--')
axs[1].set_title('cos(x)')

# Plotting tan(x) on the third subplot (the rightmost subplot)
axs[2].plot(x, y3, 'g--')
axs[2].set_title('tan(x)')

plt.show()
#'r', 'b', and 'g': These are color codes for red, blue, and green, respectively.
#They specify the color of the line in the plot.
#In this case, 'r' stands for red, 'b' stands for blue, and 'g' stands for green.
#'--': This is the line style code.
#The '--' stands for a dashed line.
#It defines how the line connecting the data points will appear in the plot.
#In this case, a dashed line will be used for all three plots.
```

```python
# plt -> implicit figure and axes creation
# Define an array of 100 equally spaced points from 0 to 10
x = np.linspace(0, 10, 100)
# Plot a line graph with x as x-axis values and x as y-axis values, using a solid line ('-')
plt.plot(x, x, '-', label = 'linear')
# Plot another line graph with x as x-axis values and x^2 as y-axis values, using a dashed line ('--')
plt.plot(x, x**2, '--', label = 'quadratic')
# Plot another line graph with x as x-axis values and x^3 as y-axis values, using circle markers and a dotted line (':')
plt.plot(x, x**3, 'o:', label = 'cubic' )
plt.xlabel('my x label')
plt.xlabel('my y label')
plt.title('simple plot with implicit graph setting')
plt.show()
```

```python
# explicit
x = np.linspace(-10, 10, 25)
fig, ax = plt.subplots() # fig and ax are objects
ax.plot(x, x,'g-', label='linear')
ax.plot(x, x**2,'r--', label='quadratic')
ax.plot(x, x**3,'o:', label='quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Simple Plot")
ax.legend()
plt.show()
```

 The ax.legend() function in Matplotlib is used to display a legend on the plot. A legend is a small box that provides information about the different elements (lines, markers, etc.) present in the plot, along with their corresponding labels.

 Scatter plots are used to visualize the relationship between two variables by plotting individual data points on a 2D plane, where each point represents one data instance with its x and y values.
 syntax ax.scatter(x, y, s=None, c=None, marker=None, cmap=None, ...)


```python
# using dictionary for graph
data = {
    #array containing values from 0 to 49 
     "a": np.arange(50),
     #array containing 50 random integers between 0 and 49 
     "b": np.random.randint(0, 50, 50),
     #array containing 50 random numbers from a standard normal distribution 
     "c": np.random.randn(50),
     #array containing values from 0 to 49, each multiplied by 10 and then added with random values from a standard normal distribution
     "d": np.arange(50) + 10 * np.random.randn(50)
 }
# Update 'd' in the 'data' dictionary to contain the absolute values multiplied by 10
data['d'] = np.abs(data['d']) * 10

fig, ax = plt.subplots(figsize=(10,3))
# The x-axis values will be taken from 'a', the y-axis values from 'b', and the color from 'c'.
# The size of the markers will be determined by 'd', and the color map 'tab20' will be used for color mapping.
ax.scatter('a', 'b', c='c', s='d' , data= data, cmap= 'tab20' )
ax.set_xlabel('entry a')
ax.set_ylabel("entry b")
ax.set_title('Scatter plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
data = {
    'names' : ['Bob', 'Jessica', 'Mary', 'John', 'Mel'],
    'english': [76, 95, 77, 78, 39],
    'math': [73, 86, 57, 18, 29],
    'hindi': [71, 81, 57, 78, 39],
}
```

```python
def my_custom_plotter(ax, x_value, y_value, param_dict): #custom plotting function called my_custom_plotter, which creates bar plots for the specified data. 
    out = ax.bar(x_value, y_value, **param_dict)#param_dict: A dictionary containing additional parameters to customize the bar plot appearance.
    ax.set_title("Visualizing marks")
    ax.set_xlabel("Marks")
    ax.set_ylabel("Count")
    return out
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 3))
my_custom_plotter(axs[0], 'names', 'english', {'data':data, 'color':'red','alpha':.3})
my_custom_plotter(axs[1], 'names', 'hindi', {'data':data, 'color':'green', 'alpha':.3})
my_custom_plotter(axs[2], 'names', 'math', {'data':data})
#The alpha parameter takes a floating-point value between 0 and 1,
# where 0 means completely transparent (invisible) and 1 means fully opaque (completely solid).
```
